```python
import numpy as np
from numpy import genfromtxt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib notebook
import sys; sys.path.append('/home'); import Conn350Lib as CL
```

```python
data = np.loadtxt ('EnROADS_with_Carbon_Price.txt') # This file has spaces as delimiters
# data =genfromtxt('En-Roads Baselinecsv.csv', delimiter=",")

# Getting the time array
time=data[:,0]

# Converting GtCO2/year to GtC/year (baseline and current scenario)
baseline = data[:,1]*0.27
current_sec = data[:,2]*0.27

# Plotting
plt.figure()
plt.plot(time,baseline, label="baseline")
plt.plot(time,current_sec, label = 'Current Scenario')
plt.xlabel('year')
plt.ylabel('emissions (gt C/year)')
plt.title('CO2 equivalent emissions by year')
plt.legend()
plt.grid(True)
```

```python
# This makes a hybrid emissions scenario file from the array "current_sec"
t_start = 1750
t_stop = 2100
filename = 'eps_hybrid.hdf5'
CL.MakeHybridEmissionScenario(t_start,t_stop,time,current_sec,filename,reportflag=False,plotflag=True)
```
